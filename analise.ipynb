```python
import geopandas as gpd
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
camada = gpd.read_file(r"LOTES_SANTO_INACIO\LOTES_SANTOINACIO_.shp",crs = 31984)
nucleo = 'SANTO INACIO DO PIAUI'
cidade = 'SANTO INACIO DO PIAUI'
uf = 'PI'
zona_utm = '24'
```

```python
def preencher_campos(camada,nucleo = 'SANTO INACIO DO PIAUI',cidade = 'SANTO INACIO DO PIAUI',uf = 'PI',zona_utm = '24'):
    camada['NUCLEO'] = nucleo
    camada['CIDADE'] = cidade
    camada['UF'] = uf
    camada['ZONA_UTM'] = zona_utm
    camada[['QUADRA', 'LOTE']] = camada[['QUADRA', 'LOTE']].fillna(0)
    camada['AREA_M2'] = camada.area
    camada['PERIMETRO'] = camada.length  

preencher_campos(camada,nucleo,cidade,uf,zona_utm)
```

```python
camada
```

```python
duplicados = camada[camada.duplicated(subset='geometry',keep=False)]
duplicados_nulos = duplicados[ (duplicados['LOTE'].isna())]

# Remover registros duplicados onde o campo 'LOTE' é nulo
camada = camada.drop(duplicados_nulos.index)
```

```python
# Transformar cada polígono em sua linha de contorno
linhas = camada.copy()
linhas['geometry'] = camada.boundary
```

```python
# Selecionar elementos que possuem valores preenchidos na coluna 'QUADRA'
gdf_preenchidos = linhas[linhas['OBSERVACAO'].notnull()  | (linhas['TIPO'] == 'NUCLEO')]
gdf_n_preenchidos = linhas[linhas['OBSERVACAO'].isnull() | (linhas['TIPO'] == 'NUCLEO')]
gdf_n_preenchidos
```
